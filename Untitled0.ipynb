```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. CSV-Daten einlesen
# Da die Daten im PDF-Format vorliegen, müssen wir sie manuell als DataFrame erstellen
data = {
    'Kunde': ['Max', 'Erika', 'Paul', 'Sofia', 'Lucas', 'Lena', 'Tom', 'Anna', 'Johannes', 'Klara',
              'Leon', 'Emma', 'Jakob', 'Marie', 'Felix', 'Nina', 'Oliver', 'Mia', 'Jonas', 'Laura'],
    'Nachname': ['Mustermann', 'Musterfrau', 'Beispiel', 'Muster', 'Test', 'Schulz', 'Fischer',
                 'Schmidt', 'Lange', 'Meyer', 'Bauer', 'Wolf', 'Weber', 'Klein', 'Braun',
                 'Hoffmann', 'Neumann', 'Krüger', 'Weiß', 'Keller'],
    'Alter': [30, 25, 45, 35, 28, 41, 39, 24, 50, 32, 27, 29, 38, 43, 36, 40, 33, 26, 44, 31],
    'Geschlecht': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'Stadt': ['Berlin', 'Hamburg', 'München', 'Berlin', 'Stuttgart', 'Hamburg', 'Frankfurt', 'Köln',
              'München', 'Düsseldorf', 'Hannover', 'Bremen', 'Berlin', 'Stuttgart', 'Frankfurt',
              'Dortmund', 'Köln', 'Berlin', 'Düsseldorf', 'München'],
    'Kaufbetrag': [150, 200, 120, 300, 175, 220, 210, 160, 280, 190, 140, 130, 270, 310, 180, 240,
                   220, 210, 250, 130]
}

df = pd.DataFrame(data)

# 2. Datenbereinigung
# In diesem Fall gibt es keine fehlenden Werte

# 3. Einfache Datenanalyse
# Durchschnittsalter der Kunden
durchschnittsalter = df['Alter'].mean()
print(f"Durchschnittsalter: {durchschnittsalter:.2f} Jahre")

# Gesamtbetrag der Käufe
gesamtbetrag = df['Kaufbetrag'].sum()
print(f"Gesamtbetrag der Käufe: {gesamtbetrag} €")

# Anzahl der Kunden pro Stadt
kunden_pro_stadt = df['Stadt'].value_counts()
print("\nAnzahl der Kunden pro Stadt:")
print(kunden_pro_stadt)

# 4. Visualisierung: Kaufbeträge nach Geschlecht
plt.figure(figsize=(10, 6))
sns.barplot(x='Geschlecht', y='Kaufbetrag', data=df, estimator=sum, ci=None)
plt.title('Gesamtkaufbetrag nach Geschlecht')
plt.xlabel('Geschlecht')
plt.ylabel('Gesamtkaufbetrag (€)')
plt.show()
```
